# Equations

```python
# a + b ---> Addition
# a - b ---> Subtraction
# a * b ---> Multiplication
# a / b ---> True Division (has decimals)
# a // b ---> Floor Division (removing decimals, rounding)
# a % b ---> Modulus (integer remainder)
# a ** b ---> Exponents
# -a ---> Negative
```

# Load Necessary Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Basic Line Graph

```python
x = [0, 1, 2, 3, 4, 5]
y = [0, 2, 4, 6, 8, 10]

plt.figure(figsize=(10,7))

plt.plot(x, y, label='2x', color='red', linewidth=2, marker='.', markersize=10, markeredgecolor='black', linestyle='--')

x2 = np.arange(0,5,0.5)
plt.plot(x2, x2**2, 'blue', label='x^2')

plt.title("My First Graph", fontdict={'fontname': 'Times New Roman', 'fontsize': 23})

plt.xlabel("x-axis", fontdict={'fontname': 'Times New Roman', 'fontsize': 23})
plt.ylabel("y-axis", fontdict={'fontname': 'Times New Roman', 'fontsize': 23})

plt.legend()

plt.savefig('myfirstgraph.png')

plt.show()
```

# Basic Bar Graph

```python
labels = ['A', 'B', 'C']

values = [1, 2, 3]

bars = plt.bar(labels, values)

plt.figure(figsize=(10,7))

plt.show()
```
